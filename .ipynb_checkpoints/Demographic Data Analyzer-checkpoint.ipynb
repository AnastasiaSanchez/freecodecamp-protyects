```python
#How many people of each race are represented in this dataset? This should be a Pandas series with race names as the index labels.
race_count = df.value_counts('race')
```

```python
#What is the average age of men?
average_age_men= round(df.loc[df['sex'].str.contains('Male'), 'age'].mean(),1)
```

```python
#What is the percentage of people who have a Bachelor's degree?
bacherlos= len(df.loc[df['education'].str.contains('Bachelors')])
total_df= len(df)
p_bachelors= round(bacherlos / total_df * 100,1)
```

```python
#What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?
h_education=df['education'].isin(['Bachelors','Masters','Doctorate'])
h_salary= df['salary'] =='>50K'
higher_education_rich= round((h_education & h_salary).sum() / h_education.sum() * 100, 1)
```

```python
#What percentage of people without advanced education make more than 50K?
l_education =df[~df['education'].isin(['Bachelors','Masters','Doctorate'])]
l_salary= len(l_education[l_education.salary == '>50K'])
lower_education_rich = round(l_salary /len(l_education) * 100,1)
```

```python
#What is the minimum number of hours a person works per week?
min_hours= df['hours-per-week'].min()
```

```python
#What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?
num_min_workers= df['hours-per-week']== min_hours
salary_p = df['salary'] =='>50K'
minhours_m50K= round((num_min_workers & salary_p).sum() / num_min_workers.sum() * 100,1)
```

```python
#What country has the highest percentage of people that earn >50K and what is that percentage?
a= df['native-country'][salary_p].value_counts()
b= df['native-country'].value_counts()
p_country = round(a/b*100,1).sort_values(ascending=False)
```

```python
#Identify the most popular occupation for those who earn >50K in India.
c= df['native-country'].isin (['India'])
top_occupation_India= df['occupation'][salary_p][c].value_counts().index[0]
```
